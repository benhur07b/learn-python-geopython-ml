# Getting started

## Setting up your environment

### uv approach
1. Download the [``pyproject.toml``](pyproject.toml) file or download the full repository [here](https://github.com/benhur07b/learn-python-geopython-ml).
    - If you just download the file, make sure to rename it to ``pyproject.toml``.
2. Go to the directory that has the ```pyproject.toml``` file.
3. Use [uv](https://docs.astral.sh) to install create your environment by:

```sh
uv sync
```

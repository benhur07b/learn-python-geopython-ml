# geoai

The **geoai** Python package developed by Qiusheng Wu (often referred to as *geoai-py*) is designed to bridge geospatial data processing and machine learning within a unified Python workflow. Built to complement the broader geospatial Python ecosystem (e.g., raster processing, vector analysis, and scientific computing), the package provides tools that simplify the preparation, modeling, and evaluation of spatial data for GeoAI applications. It is particularly useful for remote sensing, land cover classification, environmental modeling, and other spatial prediction tasks where machine learning is applied to geographic datasets. It integrates popular AI frameworks including PyTorch, Transformers, PyTorch Segmentation Models, and specialized geospatial libraries like torchange, enabling users to perform complex geospatial analyses with minimal code.

## Setting Up GeoAI (geoai-py)

In this section, we focus on setting up a clean and reproducible Python environment for working with **geoai**. Proper environment configuration ensures compatibility with dependencies such as NumPy, scikit-learn, and raster processing libraries. Establishing a well-structured environment is a critical first step in any GeoAI workflow, as it supports reproducibility, portability, and efficient collaboration—especially in training and research contexts.


### Prerequisites

GeoAI requires the following:
 - Python 3.10 or above
 - The required dependencies will be installed automatically

### Recommended Installation Methods

#### 1. Using pip
The simplest way to install the latest stable release of GeoAI is via pip:
> pip install geoai-py

To install GeoAI with all optional dependencies for additional features:
> pip install "geoai-py[all]"

#### 2. Using uv
To install the latest stable release of GeoAI with uv, a faster alternative to pip:
> uv pip install geoai-py

#### 3. Using conda
For Anaconda/Miniconda users, we recommend installing GeoAI via conda-forge, which handles dependencies like GDAL more elegantly:
> conda install -c conda-forge geoai

#### 4. Using mamba
Mamba provides faster dependency resolution compared to conda. This is especially useful for large packages like GeoAI:
> conda create -n geo python=3.12
> conda activate geo
> conda install -c conda-forge mamba
> mamba install -c conda-forge geoai

## Advanced Installation Options

### GPU Support
To enable GPU acceleration for deep learning models (requires NVIDIA GPU):
>mamba install -c conda-forge geoai "pytorch=*=cuda*"

This will install the appropriate PyTorch version with CUDA support. If you run into issues with the ipympl package, you can install it using the following command:
> mamba install -c conda-forge geoai "pytorch=*=cuda*" jupyterlab ipympl

If you encounter issues with the sqlite package, you can update it using the following command:
>mamba update -c conda-forge sqlite


### Notes for Windows Users
If you use mamba to install geoai, you may not have the latest version of torchgeo, which may cause issues when importing geoai. To fix this, you can install the latest version of torchgeo using the following command:
> pip install -U torchgeo

## Verifying Installation

To verify your installation, run:

> import geoai

> print(geoai.__version__)

Please look at the installation guide for more information on how to install the geoai package: https://opengeoai.org/installation/#upgrading

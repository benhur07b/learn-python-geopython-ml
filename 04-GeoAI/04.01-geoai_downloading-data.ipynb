# 3. GeoAI – Downloading Data

Data acquisition is a fundamental step in any GeoAI workflow. Before building models or performing spatial analysis, we need reliable and well-structured geospatial datasets such as satellite imagery, raster layers, vector boundaries, or labeled training samples. The geoai package provides convenient utilities to help users access and organize example datasets for experimentation and training.

In this section, we demonstrate how to download and manage datasets required for GeoAI workflows. Establishing a clear data structure at this stage ensures smoother preprocessing, feature engineering, and model training in the succeeding sections.


## Define Bounding Box

Define the geographic extent (longitude and latitude) for data downloads.

```python
import leafmap

m = leafmap.Map(center=[14.8480,120.5398], zoom=10)
m.add_basemap("Esri.WorldImagery")
m
```

Use the drawing tools to draw a rectangle on the map. If no rectangle is drawn, the default ROI will be used.

```python
bbox = m.user_roi_bounds()
if bbox is None:
    bbox = (120.2191088809948241, 14.6065361055562164, 120.8871514225257187, 15.0715676631444939)
```

```python
bbox
```

## Download Building Data

Retrieve building footprint data in GeoJSON format within the bounding box. The verbose flag provides detailed output.


```python
# Download buildings
from geoai.download import (
    download_overture_buildings,
)

data_file = download_overture_buildings(
    bbox=bbox,
    output="data/geoai/download/buildings.geojson",
)
```

## Extract Building Statistics

If the building data file is successfully downloaded, extract and display relevant statistics such as area, count, and footprint details.

```python
from geoai.download import (
    extract_building_stats,
)

stats = extract_building_stats(data_file)
print(stats)
```

```python
# Visualize Datasets
# Might take some time to load

m = leafmap.Map(center=[14.8480,120.5398], zoom=10) # change the center and zoom level as required by your data
m.add_basemap("Esri.WorldImagery")
m.add_geojson("data/geoai/download/buildings.geojson", layer_name="Buildings")
m
```

# Download Sentinel-2 Data from Planetary Computer

You can also download Sentinel-2 data from the Planetary Computer using the geoai package.

```python
# Import the geoai package
import geoai

# Search data
# Go to the Planetary Computer and search for the data you are interested in. Copy the STAC item link and paste it to the code below.

# Example for Sentinel-2 item
item_url = "https://planetarycomputer.microsoft.com/api/stac/v1/collections/sentinel-2-l2a/items/S2B_MSIL2A_20250804T021529_R003_T51PTS_20250804T041948"

# Download the data
# Specify which bands to download (Sentinel-2 bands)
bands_to_download = [
    "B01",
    "B02",
    "B03",
    "B04",
    "B05",
    "B06",
    "B07",
    "B08",
    "B8A",
    "B09",
    "B11",
    "B12",
    "AOT",
    "WVP",
    "SCL",
]

# Create a directory for the downloaded bands
output_directory = "data/geoai/download/sentinel2_bands"

# Download the bands, save them to the output directory, and create a merged GeoTIFF.
# The download process may take a while. Please be patient.
downloaded_bands = geoai.download_pc_stac_item(
    item_url=item_url,
    bands=bands_to_download,
    output_dir=output_directory,
    show_progress=True,
    merge_bands=True,
    merged_filename="sentinel2_all_bands.tif",
    overwrite=False,  # Skip files that already exist
    cell_size=10,  # Resample all bands to 10m resolution
)
```

```python
# Print the paths to the downloaded files
for band, path in downloaded_bands.items():
    print(f"Downloaded {band}: {path}")
```

```python
# Visualizing the data
# Raster might take some time to load
raster = "data/geoai/download/sentinel2_bands/sentinel2_all_bands.tif"
geoai.view_raster(raster, indexes=[4,3,2], visible=True, zoom_to_layer=True)
```

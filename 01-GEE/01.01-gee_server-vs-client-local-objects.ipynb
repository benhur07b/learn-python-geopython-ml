# Server-side vs Client-side/Local Objects

## Server-side
- stored on GEE's servers (not on your local machine)
- any variable created with ``ee.*`` is a server-side object
- can't be used directly with non-``ee.*`` variables or functions

## Client-side or local
- stored in your local machine
- regular Python variable

```python
import ee
```

```python
# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize(project='your-gee-project-id-here') #replace your-gee-project w/ your actual GEE project name
```

```python
# Define some local (client-side) variables
local_num1 = 100
local_num2 = 200

# Define some server-side (ee.*) variables with the same value
server_num1 = ee.Number(local_num1)
server_num2 = ee.Number(local_num2)

# Get the sum of the local variables then display it
local_sum = local_num1 + local_num2
print(local_sum)
print(type(local_sum))
```

```python
# Checking what class the server-side variables are may give us an idea
print(type(server_num1))

# Use the add method of the ee.Number class (https://developers.google.com/earth-engine/apidocs/ee-number-add)
server_sum = server_num1.add(server_num2)

# Use .getInfo()
print(server_sum.getInfo())
print(type(server_sum.getInfo()))
```

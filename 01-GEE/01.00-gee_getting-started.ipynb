## Authentication and initialization

The first step for you to be able to use GEE in Python is by authenticating/linking your GEE project.

```python
import ee
```

```python
# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize(project='your-gee-project-id-here') #replace your-gee-project w/ your actual GEE project name
```
